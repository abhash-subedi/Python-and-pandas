```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
ROOT_DIR = Path.cwd()
ss = ROOT_DIR/'data'/'superstoreSales.csv'
df = pd.read_csv(ss)
```

Group total Sales by Region and Product Category , then unstack so categories become
columns

```python
ds = df.groupby(['Region', 'Product Category'])['Sales'].sum()
print(ds)
```

From that grouped Series, use .xs to pull the sales for "Technology" across all regions.

```python
ds.xs('Technology', level='Product Category')
```

```python
df.set_index('Order ID')
```

```python
num_col = df.select_dtypes(include=[np.number])

nmedian = num_col.median()

df[num_col.columns] = num_col.mask(num_col < 0, nmedian, axis=1)
print(df.to_string())
```
