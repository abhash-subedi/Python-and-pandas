```python
import pandas as pd
from pathlib import Path
```

```python
ROOT_DIR = Path.cwd()
ss = ROOT_DIR/'data'/'superstoreSales.csv'
df = pd.read_csv(ss, encoding='latin1')
df.info()
```

Use named aggregation to get total, average, and count of Sales per Region in one call.

```python
grouped_df = df.groupby('Region').agg(total=('Sales', 'sum'), avg=('Sales', 'mean'), count=('Sales', 'count'))
grouped_df
```

Add a share_of_region column giving each order's fraction of its region's total sales (use
transform ).

```python
df['share_of_region'] = df['Sales']/df.groupby('Region')['Sales'].transform('sum')
df
```

Keep only regions whose total sales exceed 1,000,000 (use filter ), and list which
survive.

I am using a subset of the data, none of the regions as a part of this data set have sales that surpass the requirement to show up

```python
big_region = df.groupby('Region').filter(lambda x: x['Sales'].sum() > 1000000)
surving_region = big_region.groupby('Region')['Sales'].sum()
surving_region
```

Find the top 3 orders by Sales within each Region

```python
top3 = (df.groupby('Region')).apply(lambda x: x.nlargest(3, 'Sales'))
top3
```
