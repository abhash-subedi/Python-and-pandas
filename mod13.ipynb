```python
import pandas as pd
from pathlib import Path
```

```python
ROOT_DIR = Path.cwd()
ss = ROOT_DIR/'data'/'superstoreSales.csv'
df = pd.read_csv(ss, encoding='latin1')
df.info()
```

Use named aggregation to get total, average, and count of Sales per Region in one call.

```python
dg = df.groupby('Region').agg(total=('Sales', 'sum'), avg=('Sales', 'mean'), count=('Sales', 'count'))
dg
```

Add a share_of_region column giving each order's fraction of its region's total sales (use
transform ).

```python
df['share_of_region'] = df['Sales']/df.groupby('Region')['Sales'].transform('sum')
df
```
