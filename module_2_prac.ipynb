```python
#question asked for fizzbuzz to be printed for numbers that are divisible
#by both 3 and 5 and fizz if divisible by 3 and buzz if divisible by 5
nums = range(1,101)
new_list = ['fizzbuzz' if i%3==0 and i%5==0 else 'fizz' if i%3==0
            else 'buzz' if i%5==0 else i for i in nums]
print(new_list)
```

```python
# question asked for code that makes fibonacci sequence
number = int(input('please enter number: '))

def fib(n:int):
    
        if n ==0:
            return 0
        if n ==1:
            return 1
        return fib(n-2) + fib(n-1)
print(fib(number))
```

```python
# question asked to generate a number guessing game
import random as rand

secret = rand.randint(1,100)

guess_correct = False
while not guess_correct:
    guess = input('enter number guess: ')
    guess = int(guess)
    if secret>guess:
        print('the number is higher')
    elif secret<guess:
        print('the number is lower')
    else:
        print('you guessed correct')
        guess_correct = True
```

```python
# questioned for code that takes a number as input and returns true if prime false if not
def prime_checker(n:int)->bool:
    if n <=1:
        return False
    if n<=3:
        return True
    if n%2==0 or n%3==0:
        return False
    i = 5
    while i *i<=n:
        if n%i==0 or n%(i+2):
            return False
        i+=1
    return True

prime_checker(83)
```
