```python
import pandas as pd
from pathlib import Path
```

```python
ROOT_DIR = Path.cwd()
ss = ROOT_DIR/'data'/'superstoreSales.csv'
customer = ROOT_DIR/'data'/'customers.csv'
orders = ROOT_DIR/'data'/'orders.csv'

dfc = pd.read_csv(customer)
dfo = pd.read_csv(orders)
```

Merge orders with customers using how="left" and indicator=True ; check whether any
orders had no matching customer.

```python
pd.merge(dfc, dfo, how='left', indicator=True)
```

Re-run the merge with validate="many_to_one" and confirm it passes (many orders per
one customer)

```python

merged_df = pd.merge(dfc, dfo, how='left', indicator=True, validate='many_to_one')
merged_df
```
