```python
# calculator to determine the price after tax is added to item using qutity and price
price, num = input('enter price and quantity').split()

price = int(price)
num = int(num)

subtotal = price*num
tax = price*.13
total_precise = subtotal+tax

total = round(total_precise, 2)
print(total)
```
