```python
import pandas as pd
from pathlib import Path
```

```python
ROOT_DIR = Path.cwd()
titanic = ROOT_DIR/'data'/'titanic.csv'
df = pd.read_csv(titanic)
df.head()
```

Extract each passenger's title ( Mr , Mrs , Miss , Master , …) from the Name column into a
new Title column, then count how many of each.

```python
df['first']=df['Name'].str.split(',').str.get(1).str.strip()
df['title'] = df['first'].str.split('.').str.get(0)
print(df['title'].value_counts())
df
```

Create a boolean column flagging passengers whose name contains "Miss" .

```python
df['contains_miss'] = df['title'].str.lower().str.strip().str.contains('miss')
df
```

Split Name on the comma and keep just the surname (the part before it).

```python
df['surname'] = df['Name'].str.split(',').str.get(0)
df
```
